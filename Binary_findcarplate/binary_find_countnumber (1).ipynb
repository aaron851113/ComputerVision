```python
import cv2
import numpy as np  
import matplotlib.pyplot as plt
img = cv2.imread("carplate.jpg")
im_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(im_gray, cmap="gray")
plt.show()
```

```python
retval, im_at_fixed = cv2.threshold(im_gray, 180, 150, cv2.THRESH_BINARY) 
plt.title("Fixed Thresholding")
plt.imshow(im_at_fixed, cmap = 'gray') 
plt.show()
```

```python
contourimg = img
contours, hierarchy = cv2.findContours(im_at_fixed,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)  
c=0
cnts = sorted(contours, key=cv2.contourArea, reverse=True)
for contour in cnts[:1000]:
    area = cv2.contourArea(contour)
    if 800< area < 5000:
        c=c+1
        cv2.drawContours(contourimg, [contour], -1, (0, 255, 0), 3)
        cv2.putText(img,str(c),(contour[0][0][0],contour[0][0][1]),cv2.FONT_HERSHEY_SIMPLEX,1, (255, 0, 0), 2, cv2.LINE_AA)
plt.imshow(contourimg)
plt.show()
cv2.imwrite('final.jpg',contourimg)
```
