```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('photo.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot()
plt.title('Gray Photo')
plt.imshow(image,cmap = 'gray')
cv2.imwrite('photo_gray.jpg', image)
```

```python
h,w = image.shape
print(h,w)
output = np.zeros((h,w))
```

```python
Gx = np.array([[-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1]])
Gy = np.array([[-1, -2, -1],
               [0, 0, 0],
               [1, 2, 1]])


gh,gw = Gx.shape
# gc = kernel center
gc = gh//2
# kernal start from 
k_s = (gh//2)*-1
```

```python
for i in range(1, w - 1):
    for j in range(1, h - 1):
        conv_x_out = 0
        conv_y_out = 0
        for kh in range(k_s,abs(k_s)+1) :
            for kw in range(k_s,abs(k_s)+1) :
                conv_x_out += (image[i+kh][j+kw]*Gx[gc+kh][gc+kw])
                conv_y_out += (image[i+kh][j+kw]*Gy[gc+kh][gc+kw])
        output[i][j]= min(255, np.sqrt(conv_x_out**2 + conv_y_out**2))
```

```python
plt.imshow(output,cmap = 'gray')
cv2.imwrite('output.jpg', output)
```
